### The following snippet allows the user the ability to hide and show specific elements, color by chain, and label chains, we can modify the py3Dmol code accordingly.

This structure is Cetuximab, a therapeutic antibody in development for cancer treatment. Antibodies are composed of two heavy chains and two light chains; this particular construct is known as a Fab fragment and contains one full light chain (chain A) and the N-terminal half of one heavy chain (chain B). At one end of the antibody are six loops known as the “complementarity determining regions” which bind a particular antigen. 

```python
import py3Dmol

def view_pdb_with_controls(pdb_id):
    # Create a py3Dmol viewer
    pdb_viewer = py3Dmol.view(query=f'pdb:{pdb_id}')
    
    # Hide everything initially
    pdb_viewer.setStyle({}, {'hide': True})
    
    # Show cartoon representation of the structure
    pdb_viewer.setStyle({'cartoon': {'color': 'spectrum'}})
    
    # Color by chain
    pdb_viewer.setStyle({'chain': 'A'}, {'cartoon': {'color': 'red'}})
    pdb_viewer.setStyle({'chain': 'B'}, {'cartoon': {'color': 'green'}})
    
    # Label chains
    pdb_viewer.addLabel("Chain A", {'fontColor':'white', 'backgroundColor':'red', 'position':{'resn':'ALA'}, 'chain':'A'})
    pdb_viewer.addLabel("Chain B", {'fontColor':'white', 'backgroundColor':'green', 'position':{'resn':'ALA'}, 'chain':'B'})
    
    # Zoom and render the molecule
    pdb_viewer.zoomTo()
    
    # Show the molecule with interactive controls
    return pdb_viewer.show()

# View the PDB file for 1YY8
view_pdb_with_controls("1YY8")



```

```python
import pymol
from pymol import cmd

def setup_pymol():
    # Launch PyMOL without GUI
    pymol.pymol_argv = ['pymol', '-cq']  # Quiet mode, no GUI
    pymol.finish_launching()

def load_pdb(pdb_id):
    # Fetch the PDB file in PDB format
    cmd.fetch(pdb_id, name=pdb_id, type='pdb')
    cmd.hide('everything')  # Hide all for clarity
    cmd.show('cartoon', 'chain A')  # Show cartoon representation for chain A
    cmd.spectrum('count', 'rainbow', 'chain A')  # Color chain A in a rainbow spectrum

def list_atoms_chain_A():
    # Print all atoms in chain A to inspect their names
    print("Listing all atoms in chain A:")
    cmd.iterate("chain A", "print(chain, resn, resi, name)")

def measure_distance():
    # Measure distance between N of L73 and O of F21
    distance = cmd.distance('dist1', '/1YY8//A/73/N', '/1YY8//A/21/O')
    print(f"Distance between N of L73 and O of F21: {distance:.2f} Å")

def measure_angle():
    # Measure bond angle around Cβ of F21
    angle = cmd.angle('angle1', '/1YY8//A/21/CA', '/1YY8//A/21/CB', '/1YY8//A/21/CG')
    print(f"Bond angle around Cβ of F21: {angle:.2f}°")

def measure_phi_psi_chi1():
    # Measure φ, ψ, χ1 angles for F21
    # φ (phi) angle: between C of residue 20, N of residue 21, CA of residue 21, and C of residue 21
    phi = cmd.get_dihedral('chain A and resi 20 and name C', 
                           'chain A and resi 21 and name N', 
                           'chain A and resi 21 and name CA', 
                           'chain A and resi 21 and name C')
    
    # ψ (psi) angle: between N of residue 21, CA of residue 21, C of residue 21, and N of residue 22
    psi = cmd.get_dihedral('chain A and resi 21 and name N', 
                           'chain A and resi 21 and name CA', 
                           'chain A and resi 21 and name C', 
                           'chain A and resi 22 and name N')
    
    # χ1 (chi1) angle: between N of residue 21, CA of residue 21, CB of residue 21, and CG of residue 21
    chi1 = cmd.get_dihedral('chain A and resi 21 and name N', 
                            'chain A and resi 21 and name CA', 
                            'chain A and resi 21 and name CB', 
                            'chain A and resi 21 and name CG')

    print(f"φ angle of F21: {phi:.2f}°")
    print(f"ψ angle of F21: {psi:.2f}°")
    print(f"χ1 angle of F21: {chi1:.2f}°")

def add_hydrogens():
    # Add hydrogens to L73
    cmd.h_add('chain A and resi 73')
    
def measure_hbond_angle():
    # Measure the H-O-C bond angle for L73 and F21 (after adding hydrogens)
    h_o_c_angle = cmd.angle('hbond_angle', '/1YY8//A/73/H', '/1YY8//A/21/O', '/1YY8//A/21/C')
    print(f"H-O-C bond angle for backbone H-bond between L73 and F21: {h_o_c_angle:.2f}°")

# Run the functions
setup_pymol()
load_pdb('1YY8')
list_atoms_chain_A()  # Inspect atom names to confirm selections
measure_distance()
measure_angle()
#measure_phi_psi_chi1()  # This should work with simplified selections
#add_hydrogens()
#measure_hbond_angle()
```

#### We need to run the following analysis script to help us understand what non-standard residues are present in your PDB file and how to best handle them.


```python
from Bio import PDB
import warnings
from Bio import BiopythonWarning

# Suppress only specific warnings related to PDB construction
warnings.simplefilter('ignore', BiopythonWarning)

def analyze_pdb(filename):
    standard_amino_acids = {
        'ALA', 'ARG', 'ASN', 'ASP', 'CYS', 'GLN', 'GLU', 'GLY', 'HIS', 'ILE',
        'LEU', 'LYS', 'MET', 'PHE', 'PRO', 'SER', 'THR', 'TRP', 'TYR', 'VAL'
    }

    parser = PDB.PDBParser()
    structure = parser.get_structure('protein', filename)
    
    all_residues = set()
    non_standard_residues = set()
    
    for model in structure:
        for chain in model:
            for residue in chain:
                all_residues.add(residue.resname)
                if residue.resname not in standard_amino_acids:
                    non_standard_residues.add(residue.resname)
    
    print("All residues found in the PDB file:")
    print(all_residues)
    print("\nNon-standard residues found:")
    print(non_standard_residues)

# Analyze the original PDB file
analyze_pdb("1D4L.pdb")
```

#### clean up script of non-essential residues

```python
from Bio import PDB
import warnings
from Bio import BiopythonWarning

# Suppress only specific warnings related to PDB construction
warnings.simplefilter('ignore', BiopythonWarning)

def clean_pdb_advanced(input_file, output_file):
    standard_amino_acids = {
        'ALA', 'ARG', 'ASN', 'ASP', 'CYS', 'GLN', 'GLU', 'GLY', 'HIS', 'ILE',
        'LEU', 'LYS', 'MET', 'PHE', 'PRO', 'SER', 'THR', 'TRP', 'TYR', 'VAL'
    }

    parser = PDB.PDBParser()
    structure = parser.get_structure('protein', input_file)
    
    io = PDB.PDBIO()
    io.set_structure(structure)
    
    class SelectResidues(PDB.Select):
        def accept_residue(self, residue):
            # Keep standard amino acids
            if residue.resname in standard_amino_acids:
                return True
            # Keep PI9 (assumed to be important ligand/inhibitor)
            elif residue.resname == 'PI9':
                return True
            # Replace ABA with ALA (closest standard amino acid)
            elif residue.resname == 'ABA':
                residue.resname = 'ALA'
                return True
            # Remove water and sulfate ions
            else:
                return False
    
    io.save(output_file, SelectResidues())
    print(f"Cleaned PDB saved as {output_file}")

# Use the function
clean_pdb_advanced("1D4L.pdb", "1D4L_cleaned_advanced.pdb")
```

#### code that uses vector math to calculate the bond angle:

```python
import pyrosetta
from pyrosetta import Pose
import math

# Initialize PyRosetta
pyrosetta.init()


# Get the Cartesian coordinates of the N, Cα, and C atoms of residue 5
N_xyz = protease.residue(5).xyz("N")
CA_xyz = protease.residue(5).xyz("CA")
C_xyz = protease.residue(5).xyz("C")

# Calculate the vectors N->Cα and Cα->C
N_CA_vector = N_xyz - CA_xyz
CA_C_vector = C_xyz - CA_xyz

# Calculate the dot product of the two vectors
dot_product = N_CA_vector.dot(CA_C_vector)

# Calculate the magnitudes (norms) of the two vectors
N_CA_norm = N_CA_vector.norm()
CA_C_norm = CA_C_vector.norm()

# Use the dot product to calculate the cosine of the bond angle
cosine_angle = dot_product / (N_CA_norm * CA_C_norm)

# Calculate the bond angle in radians and then convert to degrees
bond_angle_radians = math.acos(cosine_angle)
bond_angle_degrees = math.degrees(bond_angle_radians)

# Print the bond angle
print(f"N-Cα-C bond angle (residue 5) using vector dot product: {bond_angle_degrees:.3f} degrees")
```

#### The following snippet will display a DataFrame with information about all residues, including whether they are protein residues, and calculate torsion angles for protein residues only.

```python
import pyrosetta
import math
import pandas as pd
from pyrosetta import Pose
from pyrosetta.rosetta.core.scoring.dssp import Dssp

def initialize_pyrosetta():
    """Initialize PyRosetta with options to handle potential issues."""
    pyrosetta.init(extra_options="-mute all -ignore_unrecognized_res -load_PDB_components false")

def load_pdb(pdb_file):
    """Load a PDB file into a PyRosetta Pose object."""
    try:
        pose = pyrosetta.pose_from_pdb(pdb_file)
        return pose
    except RuntimeError as e:
        print(f"Error loading PDB file: {e}")
        return None

def calculate_torsion_angle(pose, atom1, atom2, atom3, atom4):
    """Calculate the torsion angle between four atoms."""
    try:
        # Get Cartesian coordinates of the four atoms
        xyz1 = pose.residue(atom1[0]).xyz(atom1[1])
        xyz2 = pose.residue(atom2[0]).xyz(atom2[1])
        xyz3 = pose.residue(atom3[0]).xyz(atom3[1])
        xyz4 = pose.residue(atom4[0]).xyz(atom4[1])

        # Create vectors
        v1 = xyz2 - xyz1
        v2 = xyz3 - xyz2
        v3 = xyz4 - xyz3

        # Calculate cross products
        cross1 = v1.cross(v2)
        cross2 = v2.cross(v3)

        # Normalize the vectors
        norm1 = cross1.normalized()
        norm2 = cross2.normalized()

        # Calculate the torsion angle using atan2
        angle_radians = math.atan2(v2.norm() * v1.dot(cross2), cross1.dot(cross2))
        angle_degrees = math.degrees(angle_radians)

        return angle_degrees
    except Exception as e:
        print(f"Error calculating torsion angle: {e}")
        return None

def analyze_backbone_torsions(pose):
    """Analyze backbone torsion angles (phi, psi, omega) for all residues."""
    dssp = Dssp(pose)
    dssp.insert_ss_into_pose(pose)

    results = []
    for i in range(1, pose.total_residue() + 1):
        res = pose.residue(i)
        if res.is_protein():
            phi = pose.phi(i) if i > 1 else None
            psi = pose.psi(i) if i < pose.total_residue() else None
            omega = pose.omega(i) if i < pose.total_residue() else None
            ss = pose.secstruct(i)
        else:
            phi = psi = omega = ss = None
        
        results.append({
            'Residue': i,
            'Name': res.name(),
            'Is_Protein': res.is_protein(),
            'Phi': phi,
            'Psi': psi,
            'Omega': omega,
            'SS': ss
        })
    
    return results

def analyze_structure(pdb_file):
    """Analyze the structure and return backbone torsion angles and secondary structure."""
    initialize_pyrosetta()
    pose = load_pdb(pdb_file)
    
    if pose is None:
        return None

    return analyze_backbone_torsions(pose)

# Example usage
pdb_file = "1D4L_cleaned_advanced.pdb"  # Replace with your PDB file
results = analyze_structure(pdb_file)

if results:
    df = pd.DataFrame(results)
    display(df)

    # Print summary
    total_residues = len(results)
    protein_residues = sum(1 for r in results if r['Is_Protein'])
    non_protein_residues = total_residues - protein_residues
    print(f"\nTotal residues: {total_residues}")
    print(f"Protein residues: {protein_residues}")
    print(f"Non-protein residues: {non_protein_residues}")

    # Example: Calculate a custom torsion angle for the first two protein residues
    pose = load_pdb(pdb_file)
    protein_residues = [i for i in range(1, pose.total_residue() + 1) if pose.residue(i).is_protein()]
    if len(protein_residues) >= 2:
        res1, res2 = protein_residues[:2]
        custom_torsion = calculate_torsion_angle(pose, (res1, "N"), (res1, "CA"), (res1, "C"), (res2, "N"))
        print(f"\nCustom torsion angle (N-CA-C-N between first two protein residues): {custom_torsion:.2f} degrees")
    else:
        print("\nNot enough protein residues to calculate custom torsion angle.")
else:
    print("Failed to analyze structure.")
```

This script does the following:

Creates a 20-residue poly-alanine sequence.
Sets the phi and psi angles to ideal values for an α-helix (-57° and -47° respectively).
Saves the structure as a PDB file.
Visualizes the helix using py3Dmol.
Calculates and prints the end-to-end distance and rise per residue.

```python
import pyrosetta
import py3Dmol
import math
from pyrosetta import Pose
from pyrosetta.rosetta.numeric import xyzVector_double_t

pyrosetta.init(extra_options="-mute all")

def create_ideal_helix(sequence_length=20):
    sequence = "A" * sequence_length
    pose = pyrosetta.pose_from_sequence(sequence)
    
    # Parameters for an ideal alpha-helix
    phi = -57.8
    psi = -47.0
    omega = 180.0
    residues_per_turn = 3.6
    rise_per_residue = 1.5
    
    for i in range(1, pose.total_residue() + 1):
        pose.set_phi(i, phi)
        pose.set_psi(i, psi)
        pose.set_omega(i, omega)
        
        # Adjust the position of each residue to create the helical structure
        angle = (i - 1) * 2 * math.pi / residues_per_turn
        x = math.cos(angle) * 2.3  # Approximate radius of alpha-helix
        y = math.sin(angle) * 2.3
        z = (i - 1) * rise_per_residue
        
        # Move the CA atom of each residue
        ca_coord = xyzVector_double_t(x, y, z)
        pose.residue(i).set_xyz("CA", ca_coord)
    
    return pose

def visualize_helix(pose, width=800, height=600):
    pose.dump_pdb("ideal_alpha_helix.pdb")
    
    with open("ideal_alpha_helix.pdb", "r") as f:
        pdb_content = f.read()
    
    view = py3Dmol.view(width=width, height=height)
    view.addModel(pdb_content, "pdb")
    view.setStyle({'cartoon': {'color': 'spectrum'}})
    view.zoomTo()
    view.zoom(2.0)
    view.spin(True)
    return view

def analyze_helix(pose):
    n_term = pose.residue(1).atom("N").xyz()
    c_term = pose.residue(pose.total_residue()).atom("C").xyz()
    end_to_end_distance = n_term.distance(c_term)
    rise_per_residue = end_to_end_distance / (pose.total_residue() - 1)
    
    print(f"End-to-end distance: {end_to_end_distance:.2f} Å")
    print(f"Rise per residue: {rise_per_residue:.2f} Å")
    
    print("\nChecking hydrogen bonding pattern:")
    for i in range(1, pose.total_residue() - 4):
        n_atom = pose.residue(i).atom("N")
        o_atom = pose.residue(i+4).atom("O")
        distance = n_atom.xyz().distance(o_atom.xyz())
        print(f"Distance between N of residue {i} and O of residue {i+4}: {distance:.2f} Å")
    
    print("\nChecking phi/psi angles:")
    for i in range(1, pose.total_residue() + 1):
        phi = pose.phi(i)
        psi = pose.psi(i)
        print(f"Residue {i}: Phi = {phi:.2f}, Psi = {psi:.2f}")

# Create and analyze the ideal alpha-helix
ideal_helix = create_ideal_helix()
view = visualize_helix(ideal_helix)
view.show()

print("Manually adjusted alpha-helix has been created and visualized.")
print("PDB file saved as 'ideal_alpha_helix.pdb'")

analyze_helix(ideal_helix)

print("\nTo check the number of residues per turn, visually inspect the py3Dmol visualization.")
print("Count the number of residues it takes to make a complete turn. It should be close to 3.6.")
```
